```python
#Feature Enginnering
#Extract and clean the numeric value from string
def extract_amount(text):
    cleaned = re.sub(r'[^\d.,]','',str(text))
    cleaned = cleaned.replace(',','')
    try:
        return float(cleaned)
    except ValueError:
        return None
    
df['Amount_INR'] = df['Launched Price (India)'].apply(extract_amount)

df.head()
```
